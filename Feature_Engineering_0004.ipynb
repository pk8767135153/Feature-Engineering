#### ***Covariance*** : Covariance is a statis tical measure that examines the directional relationship between two variable. It is a way to measure how much two random variables changes together.


1.   Calculates the average of the
     product of deviations from means.
2.   * Positive : Variales move together.
     * Negative : Variable move inversely
     * Zero: No linear Reationship
3.  Sensitve to changes in variable scales. units of measuremet directly influence magnitude.
4. Provides information on direction and variability of variables.
5. We use when directional realtionship analysis. Variability assessment. Dimensionality reduction.
6. Covariance alone is not sufficient for statistical inference.

### ***Correlation*** : Correlation is a statistical term that measures how much two variables fluctuate together. It quantitatively measures the connection between two variables. Correlation may be positive, negative, or zero.

1.   Standardizes covariance by dividing by the product of standard deviations
2.   Perfect Positive (1): Variables move in perfect positive correlation. Perfect Negative (-1): Variables move in perfect negative correlation. No correlation (0): No linear relationship
3. Unitless measure: Not affected by variable scales or units of measurement.
4. Quantifies strength and direction of linear relationship.
5. Standardized relationship analysis across datasets. Multivariate analysis. Risk assessment.
6. Used in statistical tests such as t-tests, ANOVA, etc., for inference.

## **Implement Of Covariance and Correlation with python**

````python
import seaborn as sns
import pandas as pd
# Covariance
import numpy as np
````

````python
df = sns.load_dataset("healthexp")
df.drop(columns = ["Country"],inplace=True)
df.head()
````

````python
df.columns
````

````python
df.cov()
````

````python
# Pearson Correlation coeffiicient
df.corr(method = "pearson")
````

````python
#spearman rank correlation
df.corr(method = "spearman")
````
