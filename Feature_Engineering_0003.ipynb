### **Topic 02 : Nominal or One Hot Encoding :**
Nominal encoding is a technique used to transform categorical variables that have no intrinsic ordering into numerical values that can be used in maching learning models. One common method for nominal encoding is *one-hot enconding*, which creates a binary vector for each category in the variable.

<hr>
Disadvantage :  <br>
1. It will Create Sparce Matrix  and sparce matrix can generate Model Overfitting Condintion

2. One-Hot Encoding incress number of features if we increase number of feature our model we not perform well.



#### *Implement Of Nominal/OHE Encoding*

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'color':['red','blue','green','pink','red','red','green','pink']
})
```

```python
# Create an instance of One Hot Encoder
encoder = OneHotEncoder()
```

```python
encoder
```

```python
# Fit the encoder to the dataframe and transform the categorical variable.
encoded = encoder.fit_transform(df[['color']])
```

```python
encoded_df = pd.DataFrame(encoded.toarray(),columns = encoder.get_feature_names_out())
```

```python
pd.concat([df,encoded_df],axis=1)
```

#### *Ordinal Encoding*

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'size':['small','large','small','small','median','large','median','small']
})
```

```python
# create an instance of the Ordinal Encoder class and fit_transform
encoder = OrdinalEncoder(categories=[['small','median','large']])
# In Ordinal Encoding we can provide number of each value
```

```python
encoder.fit_transform(df[['size']])
```

### **Topic 04 : Target Guided Ordinal Encoding**

In is technique used to encode categorical variables based on their relationship with the target variable. This encoding technique is useful when we have a categorical variable with a large number of unique categorices and we want to use this variable as a feature in our machine learning model.

<br>

In Target Guided Ordinal Encoding. We replace each category in the categorical variable with a numerical value based  on the mean or median of the target variable for that category. This creates a monotonicrelationship between the categorical variable and the target variable which can improve the predictive power of our model.

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "city":["New York","London","Paris","Tokyo","New York","Paris"],
    "price":[200,150,300,250,180,320]
})
df
```

```python
mean_price = df.groupby('city')["price"].mean().to_dict()
```

```python
# replace each city with its mean price
df["city_encoded"] = df["city"].map(mean_price)
df
```
