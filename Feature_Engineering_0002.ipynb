* Standardiration :  Here we use the Z-Score method

#### *Implement Standardization using seaborn and numpy libary (In this method we apply formula mannualy)*

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head(2)
```

```python
# from google.colab import sheets
# sheet = sheets.InteractiveSheet(df=df)
```

```python
# We convert Series into list dtype
total_bill = list(df["total_bill"])
```

```python
mean = np.mean(total_bill)
std = np.std(total_bill)
print(f"Mean_total_bill : {mean} Standard Deviation {std}")
```

```python
# Perform normalizatio by apping z_score formula :
normalized_data = []
for i in total_bill:
  z_score = (i-mean)/std
  normalized_data.append(z_score)
print(normalized_data)
```

```python
sns.histplot(df["total_bill"])
```

```python
sns.histplot(normalized_data)
```

#### *Implement  z_Score Method by using sklearn libary*

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
df.head(2)
```

```python
# fit will find the mean and standard devation
scaler.fit(df[["total_bill","tip"]])
```

```python
# import pandas as pd
# Here we apply z_score method formula which is return me
scaler.transform(df[["total_bill","tip"]])

# Let convert value into a dataframe
df1 = pd.DataFrame(scaler.transform(df[["total_bill","tip"]]))
df1.head()
```

```python
sns.histplot(df1)
```

#### *Implement Normalization Min Max Scaler*

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("taxis")
df.head(2)
```

##### Implement Normalization Min Max Scaler By appling formula  (If there there is any mistake so sorry for mistake)

```python
lis_total = list(df["total"])

len(lis_total)
# len(lis_total) = 6433

sns.histplot(lis_total)
```

```python
x_min = min(lis_total)
x_max = max(lis_total)
normalization_data = []
for i in lis_total:
  x_scaled = (i-x_min)/(x_max - x_min)
  normalization_data.append(x_scaled)

sns.histplot(normalization_data)
```
